```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Example: create a score column (you can replace this with your real data)
np.random.seed(0)
scores = np.random.normal(loc=70, scale=10, size=100)
scores = np.clip(scores, 0, 100)

df = pd.DataFrame({"score": scores})

# Create histogram
plt.hist(df["score"], bins=10)
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.title("Histogram of Scores")
plt.show()

# Explanation
print("Explanation:")
print("The histogram shows the distribution of scores.")
print("Most scores cluster around 65–75, indicating an average near 70.")
print("Fewer students scored very low or very high marks.")
print("The distribution appears approximately normal (bell-shaped).")
```

# 07 — Data Visualization (Matplotlib + Seaborn)

## Learning goals
- Create line, bar, and histogram charts
- Use visualization for storytelling

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

sns.set_theme(style="whitegrid")

sales = pd.DataFrame({
    "month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],
    "revenue": [1200, 1350, 1280, 1500, 1650, 1720],
})

plt.figure(figsize=(8, 4))
plt.plot(sales["month"], sales["revenue"], marker="o")
plt.title("Monthly Revenue Trend")
plt.xlabel("Month")
plt.ylabel("Revenue")
plt.show()
```

```python
students = pd.DataFrame({
    "group": ["A", "A", "B", "B", "C", "C"],
    "score": [70, 75, 80, 85, 90, 95]
})

plt.figure(figsize=(6, 4))
sns.boxplot(data=students, x="group", y="score")
plt.title("Score Distribution by Group")
plt.show()
```
